# Unificar Dataframes de Customer, services e Contrats

```python
# Contar quantidade de registros por dataframe
print(f"Quantidade de registros do dataframe df_customers: {len(df_customers)}")
print(f"Quantidade de registros do dataframe df_services: {len(df_services)}")
print(f"Quantidade de registros do dataframe df_contracts: {len(df_contracts)}")

# renomear colunas
# df_services.rename(columns={'customer_id': 'customerID'}, inplace=True)

# univar dataframes de customer com services criando um terceiro dataframe
df_temp = df_customers.merge(df_services, on='customerID')

# mostrar as 5 primeiras linhas do dataframe
df_temp.info()

# df_customers.info()
# df_services.info()
```

```python
# unificar df_temp com df_contractsm usando colunas de juncao com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['customerID'], right_on=['id'])

# mostrar as 5 primeiras linhas do dataframe
df_churn_temp.info()

```

```python
# unificar os 3 dataframes em um unico dataframe, com colunas de juncao com nomes distintos
df_churn = df_customers.merge(df_services, on=['customerID']).merge(df_contracts, left_on=['customerID'], right_on=['id'])

# mostrar as 5 primeiras linhas do dataframe
df_churn.info()

```
